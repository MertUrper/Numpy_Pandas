## Gözlem ve Değişken Seçimi: loc & iloc

```python
import numpy as np
import pandas as pd
```

```python
m = np.random.randint(1,30, size = (10,3))
df = pd.DataFrame(m, columns = ["var1","var2","var3"])
df
```

```python
# loc[row_label, column_label]
# loc: tanımlandıgı şekli ile seçim yapmak için kullanılır 
```

```python
df.loc[0:3]                   # endpoint dahil ÖNEMLİ!!!!
```

```python
# iloc[row_index, column_index]
# iloc: alışık oldğumuz indexleme mantığı ile seçim yapar
```

```python
df.iloc[0:3]              
```

```python
df.iloc[:3,:2]
```

```python
df.loc[0:3,"var3"]           # verilen kurallara bağlı olarak seçim yapılma ihtiyacı varsa gözlem yada 
                             # değişken isimlendirmeleri açısından loc kullan
```

```python
df.loc[0:3,["var3"]]         # çift parantez farkı
```

```python
df.loc[:,["var3"]]            # var3 için tüm satırlar gelir
```

```python
df.loc[5:8,"var1":"var3"]
```

```python
df.loc[5:8][["var1","var3"]]
```

```python
df.İloc[0:3,"var3"]     # loc ile iloc'un en belirgin farkı, yukarıdaki işlem iloc ile yapılınca hata verdi
                        # çümkü iloc index bazlı etiket almaz          
```

```python
df.iloc[0:3]["var3"]     # verilen isimlendirmelerden bağımsız klasik index mantığı ile seçmek istersek iloc kullanılır
                         # hatanın önüne geçmek üzere değişken ayrı köşeli parantez ile oluşturuldu
```

```python
# özetle iloc index bazlı, loc ise etiket(satır ve sütun ismi) bazlı çalışır
# iloc[row_index, column_index]
# loc[row_label, column_label]
```

```python
df
```

```python
# var2 sütun ve 2. satır verisine alternatif ulaşımlar:
print(df.iloc[1,1])
print(df.iloc[1][1])
print(df.iloc[1]["var2"])
print(df.loc[1,"var2"])
print(df.loc[1]["var2"])
```

```python
m = np.random.randint(1,30, size = (10,3))
df = pd.DataFrame(m, columns = ["var1","var2","var3"])
df
```

```python
df[df['var1'].between(10,15 , inclusive = 'left')]       # 15'ler dahil olmadı
```

```python
df[~df['var1'].between(10,15)]         # " ~ " ile aralığın dışında kalanları getirir
```

```python
df.nlargest(5 , 'var2')                      # var2 için en büyük ilk 5 satırı getirdi
```

## Birleştirme İşlemleri

```python
m = np.random.randint(1,30, size = (5,3))
df1 = pd.DataFrame(m, columns = ["var1","var2","var3"])
df1
```

```python
df2 = df1 +99
df2
```

```python
pd.concat([df1,df2])       # iki kere 0 dan 4'e index yazdı yani index problemi var
                           # ayrıca index = 0 default yani satır bazında birleşme
```

```python
pd.concat([df1,df2], ignore_index = True)      # index problemi ortadan kalktı ignore_index ile
```

```python
pd.concat([df1,df2], axis = 1)            # sütun bazında birleşme
```

```python
df1 = pd.DataFrame({'anahtar': ['a', 'b', 'c'],
                    'df1_col1': range(1,4)},
                   index=['index_1', 'index_2','index_3'])
df1
```

```python
df2 = pd.DataFrame({'anahtar': ['b', 'c', 'd'],
                    'df2_col2': range(2,5)},
                   index=['index_2','index_3','index_4'])
df2
```

```python
pd.concat([df1, df2])                  # join="outer" default değeri
```

```python
pd.concat([df1, df2], join="inner")
```

```python
pd.concat([df1, df2], ignore_index=True)
```

```python
# ikiden fazla dataframe birleştirmek istersek
pd.concat([df1, df2, df2], axis=0)
```

## İleri Birleştirme İşlemleri

```python
df1 = pd.DataFrame({'anahtar': ['a', 'b', 'c'],
                    'df1_col1': range(1,4)},
                   index=['index_1', 'index_2','index_3'])
df1
```

```python
df2 = pd.DataFrame({'anahtar': ['b', 'c', 'd'],
                    'df2_col2': range(2,5)},
                   index=['index_2','index_3','index_4'])
df2
```

### on='.....'

```python
pd.merge(df1, df2, on='anahtar')
```

```python
# DataFrame ortak bir 'anahtar' değişkene sahip olduğundan, otomatik olarak buna göre birleştirilir
pd.merge(df1, df2)                
```

### left_on='anahtar1', right_on='anahtar2'

```python
df1.columns=["anahtar1","df1_col1"]
df2.columns=["anahtar2","df1_col2"]
df1
```

```python
df2
```

```python
pd.merge(df1, df2, left_on='anahtar1', right_on='anahtar2')
```

```python
pd.merge(df2, df1, left_on='anahtar2', right_on='anahtar1')        # how = "inner" default değer
```

### how{‘left’, ‘right’, ‘outer’, ‘inner’, ‘cross’}, default ‘inner’

```python
# Dikkat ettiyseniz, a ve d değerleri karşılık gelmediği için ekrana yazdırılmıyor.
# Ancak birleştirmek istersek outer seçeneğini kullanabiliriz

pd.merge(df1, df2, left_on='anahtar1', right_on='anahtar2', how='outer')
```

```python
pd.merge(df1, df2, left_on='anahtar1', right_on='anahtar2', how='left')
```

```python
pd.merge(df1, df2, left_on='anahtar1', right_on='anahtar2', how='right')
```

### suffixes : list-like, default is (“_x”, “_y”)

```python
df1 = pd.DataFrame({'anahtar': ['a', 'b', 'c'],
                    'df1_col1': range(1,4),
                    'sayi': ['bir', 'iki', 'uc']})
df1
```

```python
df2 = pd.DataFrame({'anahtar': ['b', 'c', 'd'],
                    'df2_col2': range(2,5),
                    'sayi': ['uc', 'uc', 'uc']})
df2
```

```python
# iki tane aynı isimden sütun var, seçtiğimiz üzerinden birleşme gerçekleşir diğer isim pandas tarafından otamatik değiştirilir 
pd.merge(df1, df2, on='anahtar')
```

```python
# diğer ortak sütunu biz de isimlendirebiliriz
pd.merge(df1, df2, on='anahtar', how='outer', suffixes=('_veri1', '_veri2'))
```

### on=["...","..."]

```python
# birden fazla ortak sütun varsa aşağıdaki gibi birleştirme yapabiliriz
pd.merge(df1, df2, on=['anahtar', 'sayi'])
```

```python
pd.merge(df1, df2, how='outer')         # koşulsuz birleşme
```

### right_index=True , left_index=True

```python
df1 = pd.DataFrame({'df1_col1': ['a', 'b', 'c'],
                    'df1_col2': range(1,4)})
df1
```

```python
# ortak kısım burda index kısmında yukarıdaki dataframe de sütun kısmında 
df2 = pd.DataFrame({'df2_col1': ['bir', 'iki', 'uc']},
                   index=['b', 'c', 'd'])
df2
```

```python
# Veri kümelerini df2 veri kümesinin indekslerine göre birleştirmek istiyorum.
pd.merge(df1, df2, left_on='df1_col1', right_index=True)
```

### Gruplama işlemleri

```python
df = pd.DataFrame({'gruplar':['A','B','C','A','B','C'],
                   'veri':[10,11,52,23,43,55]})
df
```

```python
df.groupby("gruplar")   # dataframe git, groupby fonk. ile gruplar değişkenini yakala ve kategori sınıflarını tut demek
```

```python
df.groupby("gruplar").mean()
```

```python
df.groupby("gruplar").sum()
```

```python
df = sns.load_dataset("planets")
df.head()
```

```python
df.groupby("method")                                # gruplama yapacak olduğu değişken method
```

```python
df.groupby("method")[["orbital_period"]].mean()      # gruplama neticesinde ilave bir değişken seçimi
```

```python
df[["method","orbital_period"]].groupby("method").mean()     # groupby 2. yöntem 
```

```python
df[["method","orbital_period"]].groupby("method").mean().sort_values(by = "orbital_period")   # sıralama
```

```python
df[["method","orbital_period"]].groupby("method", as_index = False).mean().sort_values(by = "orbital_period")
# index ilave etme
```

```python
df[["method","orbital_period","number"]].groupby(["method","number"]).mean()
```

```python
df.groupby("method")["orbital_period"].describe()   

# dataframe içerisindeki method kategorik değişkeninin sınıflarını yakaladı daha sonra 
# bu sınıfların altında orbital_period değişkennin her bir sınıf altındaki birçok değere
#(min, max, std vb.) ulaşmayı sağladı                                                     
```

```python
df.groupby("method")["orbital_period"].describe().T         # transpose alınmış hali
```
