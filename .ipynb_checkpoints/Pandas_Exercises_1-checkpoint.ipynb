```python
soccer[soccer["guaranteed_compensation"].max() == soccer["guaranteed_compensation"]]["last_name"].iloc[0]    # burdan isme ulaşıyoruz
```

#### Her takımda kaç kişinin oynadığını hesaplayın.

```python
soccer["club"].value_counts()                 
```

#### Soyadlarının içinde "son" geçen futbolcuları bulun

```python
def re_find(last_name):
    if "son" in last_name.lower():
        return True
    return False

soccer["last_name"].apply(re_find)        # apply içindeki fonksiyonu last_name'in tüm satılarına uygular 
```

```python
soccer[soccer["last_name"].apply(re_find)]         # yukarıdaki filtre işlemini soccer içerisinde sorgulayalım
```

```python
# contains ile daha pratik yöntem
soccer[soccer.last_name.str.contains("son")]
```

#### Takım ve mevki bilgisinin yer aldığı yeni bir sütun oluşturun

```python
soccer["new_info"] = soccer["club"] + "-" + soccer["position"]
```

```python
soccer.head()
```

#### En çok oyuncuya sahip takımın oyucularına yazdır

```python
# 2 farklı şekilde bu bilgiye erişebiliriz
```

```python
# 1. yöntem
soccer.club.value_counts().index[0]       # iloc[0] ise bize ilk indexin değerini verir   
```

```python
# 2. yöntem
soccer.club.mode().iloc[0]                 # en çok tekrar eden club ismini bulma (en çok oyuncuya sahiptir)
```

```python
soccer.club.value_counts().idxmax()
```

```python
soccer[soccer["club"] == soccer.club.mode().iloc[0]]
```

#### En çok oyuncuya sahip takımı getir ancak içinden maaşı ortalamanın üstünde olanları yazdır 

```python
soccer.club.mode().iloc[0]      # en çok oyuncusu olan takım 
```

```python
soccer.base_salary.mean()       # maaş ortalaması
```

```python
soccer[(soccer["club"] == soccer.club.mode().iloc[0]) & (soccer.base_salary.mean() < soccer.base_salary)]
```

#### Mevkisi sadece M-D, D-M ve F-M' den birisi olan oyuncuları getir

```python
soccer.position.value_counts()
```

```python
soccer[soccer['position'].isin(['M-D' , 'F-M' , 'D-M'])]
```

#### Mevkisi M olmayan oyuncuları getir

```python
soccer[soccer.position != "M"]
```

```python
soccer[~(soccer.position == "M")]         # yukarıdaki ile aynı sonuç
```
