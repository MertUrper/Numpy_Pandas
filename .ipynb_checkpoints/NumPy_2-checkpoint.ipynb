## Slicing ile Elemanlara Erişim (array alt kümesine erişmek )

```python
# tek boyutlu slice işlemler
```

```python
a = np.arange(20,30)
a
```

```python
a[0:3]                               # dilimleme
```

```python
a[3:]
```

```python
a[0::2]                               # 2' şer 2' şer ilerleme
```

```python
# iki boyutlu slice işlemler
```

```python
m = np.random.randint(10, size = (5,5))
m
```

```python
m[:,0]                       # ":" tüm satırları seç, 0. sütünü seç
```

```python
m[:,2]
```

```python
m[0,:]                        # 0. satır, tüm sütünları al demek
```

```python
m[1,:]
```

```python
m[0:2, 0:3]                    # hem satır hem sütün üzerinde dilimleme
```

```python
m[::,0:2]                      # m[:,:2] , m[:,0:2]  hepsi aynı sonucu verir
```

```python
m[1:3, 0:2]
```

## Alt Küme Üzerinde işlem yapma

```python
a = np.random.randint(10, size = (5,5))
a
```

```python
alt_a = a[0:3, 0:2]          # arrayimizin bir kısmını aldık
alt_a
```

```python
alt_a[0,0] = 999
alt_a[1,1] = 888              # arrayimizde değişiklik yaptık
```

```python
alt_a
```

```python
a                             # alt kümede yaptığımız değişiklik ana kümeye de yansıdı
```

```python
# bu durum bazen çok büyük arrayler de işimize yarayabilir 
# bu alt ve ana küme değişikliğinin önüne geçmek için copy işlemini kullanicez
```

```python
m = np.random.randint(10, size = (5,5))
m
```

```python
alt_b = m[0:3, 0:2].copy()
alt_b
```

```python
alt_b[0,0] = 999          # copy sayesinde alt kümede değişiklik yapsak bile ana kümeye yansimicak
```

```python
alt_b
```

```python
m                         # değişiklik olmadı
```

## Çok Boyutlu Arraylerde Iteration

```python
arr = np.arange(20).reshape(4,5)
arr
```

```python
for i in arr:
    print(i)
```

```python
for x in arr:
    for y in x:
        print(y)
```

## Çok Boyutlu Arraylerde nditer() Kullanımı

```python
# For döngüsü arrayin boyutu kadar sayıda kullanılırsa ancak array'in bütün elemanlarına tek tek erişebiliriz.
# Çok boyutlu arraylerde nditer() fonksiyonu bunun için geliştirilmiştir.
# For döngüsüne çok daha efektif ve hızlı bir şekilde tüm elemanları dönmenizi sağlar.
```

```python
arr2 = np.arange(30).reshape(-1,3,5)
arr2
```

```python
for x in np.nditer(arr2):
    print(x)
```

### Numpy array ile Armstrong Sayı kontrolü

```python
import numpy as np

num = input("Enter a number: ")
arr = np.array([int(x) for x in str(num)])
power = np.power(arr,len(arr))

if int(num) == np.sum(power):
    print(f"{int(num)} sayısı Armstrong sayıdır")
else:
    print(f"{int(num)} sayısı Armstrong sayı değildir")
```

```python
import numpy as np
num = input("Enter a number: ")
arr = np.array(list(map(int, num)))

if int(num) == np.sum(np.power(arr,len(arr))):
    print(f"{int(num)} sayısı Armstrong sayıdır")
else:
    print(f"{int(num)} sayısı Armstrong sayı değildir")
```

```python
# chat GBT'nin cevabı
import numpy as np

number = input("Sayıyı girin: ")  # Kullanıcıdan sayı alınır
digits = np.array(list(number), dtype=int)  # Sayının rakamları bir numpy array olarak alınır
n = len(digits)  # Sayının basamak sayısı hesaplanır

# Sayının Armstrong sayısı olup olmadığı kontrol edilir
if np.sum(digits ** n) == int(number):  
    print(number, "bir Armstrong sayısıdır")  
else:  
    print(number, "bir Armstrong sayısı değildir")
```
